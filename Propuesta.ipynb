# &nbsp; <font color=#0377BB> Aspectos generales</font> <a class="anchor" id="superficies"></a>

<p style="text-align: justify;">Al hablar de formas geométricas en el espacio y delimitar el campo de estudio se pueden abarcar las superficies, las cuales normalmente se ven representadas en el espacio tridimensional. Es por ello que es necesaria una definición formal, la cual se muestra a continuación.</p>

<p style="text-align: justify;"><div style="background-color: #C8E6C9 ; color:#163A18;padding: 10px; border: 1px solid green;">
<b>Definición 1.</b> Una superficie es un conjunto de puntos $(x,y,z)$ en $\mathbb{R}^3$ que satisfacen la ecuación $z=f(x,y)$ o también en la forma $F(x,y,z)=0$. </div> </p>

## &nbsp; <font color=#0377BB> Definición y clasificación de las superficies Cuádricas</font>

<p style="text-align: justify;">Dentro de la definición $1$ se encuentran las superficies cuádricas que describen fenómenos tanto matemáticos como los presentes en la realidad de manera frecuente inclusive. De forma general, se puede decir que gráficamente corresponden a una especie de plano deformado en el espacio euclidiano. Así, a continuación se brinda una definición formal de superficie cuádrica:</p>

<p style="text-align: justify;"><div style="background-color: #C8E6C9 ; color:#163A18;padding: 10px; border: 1px solid green;">
<b>Definición 2. </b>Una superficie cuádrica corresponde a una estructura geométrica descrita por una ecuación general de segundo grado en las variables $x, y, z$:

\begin{equation*}
    Ax^2+By^2+Cz^2+Dxy+Exz+Fyz+Gxyz+Hx+Iy+Jz+K=0\hspace{1.5cm}(1)
\end{equation*} </div></p>

<p style="text-align: justify;">En la ecuación anterior los términos cruzados entre las variables corresponden específicamente a las rotaciones que se pueden realizar a las superficies cuádricas. Sin embargo, dichas rotaciones no resultan pertinentes para efectos de este trabajo ya que no representan un cambio relevante en el estudio, análisis y tratamiento que se llevará a cabo en las siguientes secciones. Por esta razón, las constantes $D$, $E$, $F$ y $G$ de la ecuación $(1)$ representarán términos nulos, lo que daría por resultado una ecuación de la siguiente forma:</p>

\begin{equation*}
    Ax^2+By^2+Cz^2+Hx+Iy+Jz+K=0 \hspace{1.5cm} (2)
\end{equation*}

<p style="text-align: justify;">De esta manera, la ecuación $(2)$ pasa a ser la nueva ecuación general de las superficies cuádricas que se trabajarán. Ahora bien, mediante aplicación de la técnica de completación de cuadrados a dicha ecuación se pueden obtener las ecuaciones canónicas respectivas a cada cuádrica. En el apartado de tratamiento analítico y gráfico abarcarán estas últimas.

<p style="text-align: justify;">Generalmente, respecto a las cuádricas existen hasta el momento “17 tipos estándar”, algunas de ellas son: “paraboloide, esfera, esferoide, elipsoide, cono, hiperboloide, cilindro, cono elíptico, cilindro elíptico, hiperboloide elíptico, paraboloide elíptico, etc.” Sin embargo, esta en esta guía solo se considerará la clasificación de cuádricas en posición estándar (sin rotación) que se muestra en el siguiente cuadro.

# &nbsp;  <font color=#0377BB> Contextualización histórica y aplicabilidad de las Superficies Cuádricas</font> <a class="anchor" id="aplica"></a>

Pues bien, uno de los recursos más utilizados para justificar esta afirmación es la geometría, ya que, visualmente se pueden reconocer  y apreciar sus características elementales, inclusive para aquellos que no conocen sus propiedades axiomáticas.

<p style="text-align: justify;">Sin embargo, la geometría es fácilmente reconocible cuando se trata de figuras elementales como el círculo, el cuadrado, el rectángulo, entre otras, pero cuando se habla de las superficies cuádricas, la interpretación que le dé una persona puede ser confusa. Esto sucede debido a que estas no son muy intuitivas y perceptibles gráficamente como las elementales antes mencionadas.

<p style="text-align: justify;">Las superficies cuádricas son estructuras geométricas cada vez más utilizadas en la actualidad, pues cada una cuenta con propiedades y características propias que representan una gran funcionabilidad para el desarrollo de nuevas tecnologías, o bien, se consideran diseños atractivos para el mejoramiento de la infraestructura de alguna edificación.

<p style="text-align: justify;">Los ejemplos de aplicaciones que se abarcarán posteriormente funcionan como contraste de la presencia de las superficies cuádricas en un ámbito natural o desarrollo humano y en el matemático, pues la arquitectura y las telecomunicaciones pueden servir de puente entre la visualización física y los contenidos detrás de las ecuaciones que se trabajarán en la sección $5$ de este documento.

# &nbsp;  <font color=#0377BB> Tratamiento gráfico y analítico de las superficies cuádricas</font>

### &nbsp;  <font color=#0377BB> Curvas de nivel y trazas</font> <a class="anchor" id="trazas"></a>

<p style="text-align: justify;">Al referirnos a superficies en general y en este caso a cuádricas, a veces decimos “superficie de ecuación (explícita) $z = f(x,y)$” o “superficie de ecuación (implícita) $F(x,y,z) = 0$”. Un bosquejo de una superficie se puede hacer con un conjunto de curvas; a estas curvas se les llama ‘trazas’ o ‘cortes verticales y horizontales’.

Ahora bien, para realizar el tratamiento sobre curvas de nivel de una superficie cuadrica se puede establecer la siguiente definición:

<p style="text-align: justify;"><div style="background-color: #C8E6C9 ; color:#163A18;padding: 10px; border: 1px solid green;">
<b>Definición 3. </b> Si $S$ es una superficie en el espacio de ecuación $F(x,y,z) = 0$, todos los pares $(x,y)$ $\in  \mathbb{R}^2$ que satisfacen la ecuación $F(x,y,c) = 0$ definen una curva en el plano $XY$ (siempre y cuando este conjunto no sea vacío). A esta curva se le llama una curva de nivel de la superficie $S$. </div></p>

<p style="text-align: justify;">Por otro lado, al conjunto de curvas de nivel contenidas en la proyección $XY$ se le conoce como mapa de contorno (esto verse en la figura 7 parte (b)). Así, una ejemplicación de la definición anterior se puede encontrar en la figura $7$ parte (a), donde se dibuja una curva de nivel de un parabolide elíptico en la proyección $XY$. Asimismo, las curvas de nivel, geométricamente corresponden a la proyección sobre el plano $XY$, del corte del plano $z = c$ con la superficie $S$ (Mora, 2020).

Además, respecto al tratamiento de las trazas, se brinda la siguiente definición:

<p style="text-align: justify;"><div style="background-color: #C8E6C9 ; color:#163A18;padding: 10px; border: 1px solid green;">
<b>Definición 4. </b> Si $S$ es una superficie en el espacio de ecuación $F(x,y,z) =0$, se les llaman trazas de la superficie a las curvas (si existen):

<ul> <li>$F(x,y,c)=0$ o $z=f(x,y)$, $z=c$</li>
  <li>$F(x,c,z) =0$ o $z=f(x,c)$, $y=c$</li>
  <li>$F(c,y,z)=0$ o $z=f(c,y)$, $x=c$ </li> </ul>
    
   </div></p>

### &nbsp;  <font color=#0377BB> Cuádricas y su tratamiento algebraico</font> <a class="anchor" id="tratamiento"></a>

<p style="text-align: justify;">Como se vio en el primer apartado, la ecuación $(1)$ corresponde a la ecuación general de una superficie cuádrica, sin embargo, como no son necesarias las rotaciones que producen los términos cruzados, se tomará la ecuación $(2)$ como la ecuación general de  las cuádricas y así trabajarlas en una posición estándar.

<p style="text-align: justify;">Ahora bien, exceptuando casos degenerados, al completar cuadrados en la ecuación $(2)$ se obtienen las ecuaciones canónicas de cada una de las superficies cuádricas, las cuales cumplen con una serie de condiciones respectivas.
